```python
#importing libraries 
import numpy as np 
import matplotlib.pyplot as plt
import random
import math
```

```python
#generating a 2d lattice
n = int(input("number of vertices?")) #asks the user how many vertices they want for their lattice
def create_lattice(n): #defining a function that creates the lattice
    global lattice #makes the lattice variable a global variable that can be called when not using the function
    lattice = np.zeros((n,n)) #creates a n by n matrix of zeros, that represents an empty lattice
    print(lattice) #prints, outputs the empty lattice structure
create_lattice(n)
```

```python
#tests lattice dimensions and checks if it works 
print(lattice)
print(lattice[3,4])#prints coordinate 3,4 of the lattice 
lattice[3,4] = 4.0#sets coordinate 3,4 of the lattice equal to 4.0
print(lattice)
```

```python
#placing random discs on to the lattice


def plot_random_discs(rho_,f_a,n):
    f_b = 1 - f_a
    create_lattice(n)
    num_disks = rho_*n*n #number of disks is the number density times the area of the 2d lattice
    num_disks = int(num_disks) #sets the data type of the variable as an integer
    round(num_disks) #rounds the number of disks, to avoid decimals
    num_a = f_a*num_disks # number of A atoms in lattice is the fraction of A times the number of disks.
    num_a = int(num_a)
    round(num_a)
    num_b = num_disks - num_a # num_b is the number of disk b 
    round(num_b)
    global list_a # List_a and list_b are the lists of the coordinates of the disks in the lattice. 
    list_a = []
    global list_b
    list_b = []
    for i in range(num_a):  #iterates for number of A disks
        x = random.randint(0,n-1)#generates a random x coordinate 
        y = random.randint(0,n-1)#generates a random y coordinate
        while [x,y] in list_a:#while loop that generates new coordinates for A if the coordinate already exists in list
            x = random.randint(0,n-1)
            y = random.randint(0,n-1)
        
        list_a.append([x,y])#if not in list_a it then adds this coordinate to the list 
        lattice[x,y] = 1.0#sets value of this coordinate to 1.0, which corresponds to disk A 
    for j in range(num_b):#iterates for number of disk B 
        x = random.randint(0,n-1)#generates random x coordinate
        y = random.randint(0,n-1)#generates random y coordinates 
        while [x,y] in list_a and list_b:#While loop that generates new coordinates if coordinate already exists in list A and B
            x = random.randint(0,n-1)
            y = random.randint(0,n-1)
        
        list_b.append([x,y])
        lattice[x,y] = 2.0
    
    print(lattice) 
    





plot_random_discs(0.35,0.6,10)
print(len(list_a))
print(len(list_b))
```

```python
#CALCULATING ENERGY OF EACH CONFIGURATION OF THE SYSTEM


def boundary_conditions(x,y): #Function that enforces boundary conditions
    if x == 0:#if the x coordinate of the current position is 0
        (x-1) == n-1 #left of the x coordinate would be on the other end of the lattice
    elif x == n-1:#if x coordinate was on the ege
        (x+1) == 0#right of the edge would be back at x=0
    else:
        pass
    if y == 0:
        (y-1) == n-1
    elif y == n-1:
        (y+1) == 0
    else: 
        pass
#might have to re-do   
def calculate_energy(e_aa,e_bb,e_ab):#calculates the energy 
    num_e_aa = 0 #counts the number of e_aa interactions in lattice
    num_e_bb = 0 #counts the number of e_bb interactions in lattice 
    num_e_ab = 0 #counts the number of e_ab interactions in lattice
    for i in range(0,n-1): #for loop that iterates through every x coordinate
        x = i
        for j in range(0,n-1):#for loop that iterates through every y coordinate for every x coordinate
            y = j
            boundary_conditions(x,y) #checks if the current x,y coordinate obeys the boundary conditions
            #checks up 
            if lattice[x,y] == 1.0 and lattice[x,y-1] == 1.0: #if neighbours are type 1 
                num_e_aa += 1 #adds 1 to count of 1,1 neighbours
            elif lattice[x,y] == 1.0 and lattice[x,y-1] == 2.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x,y-1] == 1.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x,y-1] == 2.0:
                num_e_bb += 1
            else:
                pass
            #checks down
            
            if lattice[x,y] == 1.0 and lattice[x,y+1] == 1.0:
                num_e_aa += 1
            elif lattice[x,y] == 1.0 and lattice[x,y+1] == 2.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x,y+1] == 1.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x,y+1] == 2.0:
                num_e_bb += 1
            else:
                pass
            #checks left
            if lattice[x,y] == 1.0 and lattice[x-1,y] == 1.0:
                num_e_aa += 1
            elif lattice[x,y] == 1.0 and lattice[x-1,y] == 2.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x-1,y] == 1.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x-1,y] == 2.0:
                num_e_bb += 1
            else:
                pass
            #checks right
            if lattice[x,y] == 1.0 and lattice[x+1,y] == 1.0:
                num_e_aa += 1
            elif lattice[x,y] == 1.0 and lattice[x+1,y] == 2.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x+1,y] == 1.0:
                num_e_ab += 1
            elif lattice[x,y] == 2.0 and lattice[x+1,y] == 2.0:
                num_e_bb += 1
            else:
                continue
    total_e_aa = float(num_e_aa) * e_aa #total energy of e_aa
    total_e_bb = float(num_e_bb) * e_bb #total energy of e_bb
    total_e_ab = float(num_e_ab) * e_ab #total_energy of e_ab
    
    global total_energy
    total_energy = total_e_aa + total_e_bb + total_e_ab #totla combined energy of all interactions
    
    

calculate_energy(0.6,0.7,0.5)
print(total_energy)
```

```python

#develop swap move 
print(lattice)
def swap():
    x = random.randint(0,n-1)#generates coordinates
    y = random.randint(0,n-1)
    x_ = random.randint(0,n-1)#generates another set of coordinates
    y_ = random.randint(0,n-1)
    while [x,y] == [x_,y_] or lattice[x,y] == lattice[x_,y_]:#checks if the cordinates generated are the same
        #and if the coordinates corespond to the same disk type
        x_ = random.randint(0,n-1) #generates new coordinates if the coordinates were the same
        y_ = random.randint(0,n-1)
        if lattice[x,y] != lattice[x_,y_]:#if the coordinates are different then stops generating new coordinates
            break
        else:
            continue
    lattice[x,y], lattice[x_,y_] = lattice[x_,y_], lattice[x,y] #the two coordinates generated are swapped           
swap()
print(lattice)
```

```python
#Accept move
accept_ = 0 #accept variable that keeps count of accepted moves
def accept_move(T,u_old,u_new):
    global accept_#makes the variable global which can be used outside of the function
    k = 1.380649*10**-23 #boltzmann constant 
    u_nm = u_new - u_old #energy difference between before swap and after swap
    #code below is the metropolis criteria
    if u_nm < 0:
        accept_ += 1#
        print("yes")
        
    else:
        r = random.uniform(0,1)
        if math.exp(-u_nm/(k*T)) > r:
            accept_ += 1
            print("yes")
        else:
            print("no")
            
    
accept_move(300,69,54)
```

```python
def metropolis_algo(n,rho_,f_a,e_aa,e_bb,e_ab,T,n_moves): #function that runs the Metropolis montecarlo algorith
    plot_random_discs(rho_,f_a,n) #plots random disc
    dct = {1: 5., 0: 1., 2: 3.} #colours for plotting pixel chart representation of lattice
    n = [[dct[i] for i in j] for j in lattice]
    global energies
    energies = [] #list that contains all the energies after all iteration
    global trials
    trials = []#list that logs numberof trials to plot graphs 
    global acceptratio
    acceptratio = []#list that logs the acceptance ratio after every move
    plt.imshow(n, cmap='brg', vmin=1, vmax=10)
    plt.show()
    for i in range(1,n_moves): #iterates for n number of moves decided by user
        calculate_energy(e_aa,e_bb,e_ab) #calculates current energy of the system
        u_old = total_energy #energy pre-swap
        swap() #swap move 
        calculate_energy(e_aa,e_bb,e_ab)#calculates energy after swap
        u_new = total_energy #new energy after swap
        accept_move(T,u_old,u_new) #wether to accept the swap move or no?
        print(u_new)
        energies.append(u_new)
        acceptance_ratio = accept_/i
        acceptratio.append(acceptance_ratio)
        print(acceptance_ratio)
        trials.append(i)
    
    print(accept_) #the total accepted moves count
    dct = {1: 5., 0: 1., 2: 3.} #plots the pixel representation of final lattice after swap moves
    n = [[dct[i] for i in j] for j in lattice]
    
    plt.imshow(n, cmap='brg', vmin=1, vmax=10)
    plt.show()
    
metropolis_algo(50,0.95,0.3,-4,-2,2,300,100000) #running simulation 1
plt.plot(trials,energies)#plots the energy evolution of the simulation
plt.xlabel("timesteps")
plt.ylabel("energy")
plt.title("300k energy evolution")
plt.show()

plt.plot(trials,acceptratio)#plots the evolution of the acceptance ratio
plt.xlabel("timesteps")
plt.ylabel("Acceptance ratio")
plt.title("300k acceptance ratio evolution")
```
